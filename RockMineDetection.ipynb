```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

```python
def data_reading():
    df = pd.read_csv("mines-vs-rocks/sonar.all-data.csv")
    X = df[df.columns[0:60]].values
    y = df[df.columns[60]]
    
    encoder = LabelEncoder()
    encoder.fit(y)
    y = encoder.transform(y)
    Y = one_hot(y)
    return (X, Y)
```

```python
def one_hot(y):
    len_y = len(y)
    unique_labels = len(np.unique(y))
    Y = np.zeros((len_y, unique_labels))
    Y[np.arange(len_y), y] = 1
    return Y
```

```python
X, Y = data_reading()
print(X)
print(Y)
```

```python
X, Y = shuffle(X, Y, random_state = 1)
print(X, Y)
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size = 0.30)
print(train_x.shape)
print(test_x.shape)
print(train_y.shape)
print(test_y.shape)
```

```python
learning_rate = 0.001
epochs = 1000
cost_history = np.empty(shape=[1], dtype=float)
n_dim = X.shape[1]
n_class = 2
model_path = "/model"
```

```python
hidden_1 = 50
hidden_2 = 500
hidden_3 = 500

x = tf.placeholder(tf.float32, [None, n_dim])
W = tf.Variable(tf.zeros([n_dim, n_class]))
b = tf.Variable(tf.zeros([n_class]))
y_ = tf.placeholder(tf.float32, [None, n_class])
```

```python
def multilayer_parceptron(x, weights, biases):
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.sigmoid(layer_1)
    
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    layer_2 = tf.nn.sigmoid(layer_2)
    
    layer_3 = tf.add(tf.matmul(layer_2, weights['h3']), biases['b3'])
    layer_3 = tf.nn.relu(layer_3)
    
    layer_out = tf.add(tf.matmul(layer_3, weights['out']), biases['out'])
    
    return layer_out
```

```python
weights = {
    'h1': tf.Variable(tf.truncated_normal([n_dim, hidden_1])),
    'h2': tf.Variable(tf.truncated_normal([hidden_1, hidden_2])),
    'h3': tf.Variable(tf.truncated_normal([hidden_2, hidden_3])),
    'out': tf.Variable(tf.truncated_normal([hidden_3, n_class]))
}
biases = {
    'b1': tf.Variable(tf.truncated_normal([hidden_1])),
    'b2': tf.Variable(tf.truncated_normal([hidden_2])),
    'b3': tf.Variable(tf.truncated_normal([hidden_3])),
    'out': tf.Variable(tf.truncated_normal([n_class]))
}
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
y = multilayer_parceptron(x, weights, biases)

cost_function = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits = y, labels = y_))
training_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost_function)

sess = tf.Session()
sess.run(init)
```

```python
mse_history = []
accuracy_history = []

for epoch in range(epochs):
    sess.run(training_step, feed_dict={x:train_x, y_:train_y})
    cost = sess.run(cost_function, feed_dict={x:train_x, y_:train_y})
    cost_history = np.append(cost_history, cost)
    correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
    pred_y = sess.run(y, feed_dict={x:test_x})
    mse = tf.reduce_mean(tf.square(pred_y - test_y))
    mse_ = sess.run(mse)
    mse_history.append(mse_)
    
    accuracy = (sess.run(accuracy, feed_dict={x:train_x, y_:train_y}))
    accuracy_history.append(accuracy)
    
    print('epoch : ', epoch, ' | cost: ', cost, ' | MSE: ', mse_, ' | Train Accuracy: ', accuracy)
```

```python
plt.plot(mse_history, 'r')
plt.show()
```

```python
plt.plot(accuracy_history)
plt.show()
```
